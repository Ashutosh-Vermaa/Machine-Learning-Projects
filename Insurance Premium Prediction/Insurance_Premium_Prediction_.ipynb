## Model Building and Training

```python
# sns.pairplot(data, hue='expenses')
```

```python
y=data_dummy['expenses']
data_dummy=data_dummy.drop('expenses', axis=1)
```

```python
x=data_dummy
```

```python
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_scaled, y, test_size=0.2,random_state=43)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression().fit(x_train, y_train)
```

```python
pred=reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(pred, y_test))
print(mean_absolute_error(pred, y_test))
```

```python
reg.score(x_test,y_test) # the model is able to predict the results with an accuracy of 78%
```

```python
x_train.shape[1]
```

##Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'max_features': ['auto', 'sqrt', 'log2'],
    'criterion': ['mse', 'friedman_mse', 'absolute_error', 'poisson'],
    'min_samples_leaf': [1, 2, 5]
}
```

```python
model = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5)
model.fit(x_train, y_train)

# regressor = DecisionTreeRegressor(random_state=0)
# regressor.fit(x_train, y_train)
```

```python
best_params = model.best_params_
best_model = model.best_estimator_

print(best_params,best_model)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, model.predict(x_test)) #model predicts the value with 71% accuracy
```
